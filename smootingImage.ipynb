```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def showImage(img):
    plt.figure(figsize=(10,10))
    for i in range(len(img)):
        plt.subplot(1,2,i+1)
        plt.imshow(img[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()


def rgbImage(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread("./image/thai-lottery.png")
img1 = cv2.imread("./image/thai-lottery.png")

showImage([rgbImage(img), rgbImage(img1)])
```

```python
kernel = np.array([[-1,-1,-1], [-1,10,-1], [-1,-1,-1]])
img_sharpness = cv2.filter2D(img, -1, kernel)
showImage([rgbImage(img_sharpness), rgbImage(img1)])
```

```python
img_blur = cv2.blur(img, (20, 20))
showImage([rgbImage(img_blur), rgbImage(img1)])
```

```python
img_med_blur = cv2.medianBlur(img, 9)
showImage([rgbImage(img_med_blur), rgbImage(img1)])
```

```python
img_bilateral = cv2.bilateralFilter(img,9,75,75)
showImage([rgbImage(img_bilateral), rgbImage(img1)])
```
