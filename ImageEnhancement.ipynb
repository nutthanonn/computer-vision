```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def showImage(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()

def rgbImage(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread("./image/eye.png", 0)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

showImage(img_gray)
```

```python
img_gaussian = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 51, 2.5)
showImage(rgbImage(img_gaussian))
```

```python
img_enchance = cv2.equalizeHist(img)
showImage(rgbImage(img_enchance))
```
