```python
import cv2
import matplotlib.pyplot as plt


def showImage(image, color=cv2.COLOR_BGR2RGB):
    img = cv2.cvtColor(image, color)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.show()


img = cv2.imread("./image/thai-lottery.png", 1)
showImage(img, cv2.COLOR_BGR2RGB)
```

```python
#  Local Thresholding

blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2]
img_grayscale = (0 * blue) + (0 * green) + (1 * red)
showImage(img_grayscale)
```

```python
# Global Thresholding

coin_img = cv2.imread('./image/coins.png', 0)
ret,th1 = cv2.threshold(coin_img, 127, 255, cv2.THRESH_BINARY)
showImage(th1)
```

```python
# Adaptive Gaussian Thresholding

th2 = cv2.adaptiveThreshold(coin_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
showImage(th2)
```
